5.Filter Data:
1. A basic loop to filter rows based on a specific condition.
2. Print the filtered rows.

```python
returning_customer = []
for customer in df["Customer Type"]:
    if customer == "Returning":
        #print(customer)
        returning_customer.append(customer)
print(returning_customer)
```

```python
filter = df["Customer Type"] == "Returning"
print(df[filter])
#print(filter)
```
